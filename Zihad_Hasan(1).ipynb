```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('final-employee-ds.csv')
```

```python
df
```

```python
training_array = df['TrainingHours'].values.astype(float)
print(training_array)
array_mean = np.mean(training_array)
array_std = np.std(training_array)
print("The mean of Training Hour:",array_mean)
print("The standard deviation of Training Hour:",array_std)
```

## Answer to the Question: 02:

```python
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))
filtered = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]
total_salary = filtered['Salary'].sum()
print(total_salary)
```

## Answer to the Question : 03

```python
department = df['Department'].isin(['IT','Finance'])
age_condition = (df['Age'] >= 30) & (df['Age'] <= 40)
performance = df['PerformanceScore'] > 88

filtered_data = df[department & age_condition & performance]
print(filtered_data)
```

```python
sorted_salary = filtered_data.sort_values(by='Salary', ascending=False)
final_output = filtered_data[['Name', 'Department', 'Salary', 'PerformanceScore']]
print(final_output)
```

## Answer to the Question : 04

```python
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))
df['SalaryPerHour'] = ( df['Salary'] / (df['WorkHoursPerWeek']*4))
top_employees = df.sort_values(by='SalaryPerHour', ascending=False).head(5)
output = top_employees[['Name','SalaryPerHour']]
print(output)
```
